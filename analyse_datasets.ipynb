```python
#Import needed library
import os
import pandas as pd
import s3fs
import zipfile
from datasets import load_dataset
from datasets import Dataset, DatasetDict
```

```python
# Load dataset2
S3_ENDPOINT_URL = "https://" + os.environ["AWS_S3_ENDPOINT"]
fs = s3fs.S3FileSystem(client_kwargs={'endpoint_url': S3_ENDPOINT_URL})
fs.ls("civel/diffusion/hackathon-minarm-2024/AIVSAI/hack_train.csv")
PATH_IN = 'civel/diffusion/hackathon-minarm-2024/AIVSAI/hack_train.csv'
fs.download(PATH_IN, 'data/hack_train.csv')
df = pd.read_csv('data/hack_train.csv')
```

```python
#Analyse dataset2

# Checking if balanced, how many sources ect...

# Number of sources
source_counts = df['src'].value_counts()

# Number of human generated answer and AI generated answer
label_counts = df['label'].value_counts()

# How many sources are for human data and how many are for machine data  (Done because we noticed that there were many more AI sources)
sources_label_1 = df[df['label'] == 1]['src'].nunique()
sources_label_0 = df[df['label'] == 0]['src'].nunique()

# Getting the top sources for each
top_sources_label_1 = df[df['label'] == 1]['src'].value_counts().nlargest(3)
top_sources_label_0 = df[df['label'] == 0]['src'].value_counts().nlargest(3)

# Print the results
print("Number of different sources:", len(source_counts))
print("Amount of label 1:", label_counts[1])
print("Amount of label 0:", label_counts[0])
print("Number of sources with label 1:", sources_label_1)
print("Number of sources with label 0:", sources_label_0)

print("\nTop 3 sources for label 1:")
print(top_sources_label_1)

print("\nTop 3 sources for label 0:")
print(top_sources_label_0)
```

```python
# Load dataset1 - reddit and analyse 
dataset = load_dataset("Hello-SimpleAI/HC3", name="reddit_eli5")
human_answers = dataset['train']['human_answers']
chatgpt_answers = dataset['train']['chatgpt_answers']
human_answers = [string for sublist in dataset['train']['human_answers'] for string in sublist if sublist]
chatgpt_answers = [string for sublist in dataset['train']['chatgpt_answers'] for string in sublist if sublist]
data_dict = {
    "texts": human_answers + chatgpt_answers,
    "labels": [1] * len(human_answers) + [0] * len(chatgpt_answers)
}
hf_dataset = Dataset.from_dict(data_dict)

df = hf_dataset.to_pandas()

# Number of human generated answer and AI generated answer
label_counts = df['labels'].value_counts()

# Print the counts
print("Number of labels 'human':", label_counts[1])
print("Number of labels 'AI':", label_counts[0])
```
